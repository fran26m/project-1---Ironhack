```python
# 1 # standard deck: 52 cards
import time
deck = [2,3,4,5,6,7,8,9,10,10,10,10,11]*4
#len(deck)

# 2 # First 4 cards are dealt

print("the game starts, 2 cards for the player(user) and 2 cards for the dealer")
time.sleep(1)

import random
card_1_p = random.choice(deck)
card_2_p = random.choice(deck)
card_1_d = random.choice(deck)
card_2_d = random.choice(deck)
print("Player first card is: ",card_1_p)
time.sleep(1)
print("Player second card is: ",card_2_p)
time.sleep(1)
print("Dealer first card is: ",card_1_d)
time.sleep(1)
print("Dealer second card is: ",card_2_d)
time.sleep(1)


dealer_hand=[card_1_d,card_2_d]
player_hand=[card_1_p,card_2_p]
total_score_player = card_1_p + card_2_p
total_score_dealer = card_1_d + card_2_d

if total_score_player > 21:
    print("Player exceeded 21, dealer wins.")

elif total_score_dealer > 21:
    print("Dealer exceeded 21, player wins.")

else:
    


    game_state = {'total_score_player':total_score_player,'total_score_dealer':total_score_dealer, 'player_hand':player_hand,'dealer_hand':dealer_hand }
    time.sleep(2)
    print("Game state: ",game_state)
    
    
    # 3 # Next hands
    
    ###game logic #####
    
    if total_score_player < 21:
    
        player_decision = input("Enter 'hit' to play another hand or 'stand' to stop: ").lower().strip()
    
        while player_decision != 'hit' and player_decision != 'stand':
            player_decision = input("Wrong entry! Enter 'hit' to play another hand or 'stand' to stop: ").lower().strip()
    
    else:
        player_decision = 'stand'
    
    while total_score_player < 21 and player_decision == 'hit':
    
        new_card_player = random.choice(deck)
        time.sleep(1)
        print("Player new hand: ", new_card_player)
        
    
        player_hand.append(new_card_player)
        print("Player hands: ", player_hand)
    
        total_score_player = total_score_player + new_card_player
    
        #print("Player hand:", player_hand)
        print("Player score: ", total_score_player)
    
        if total_score_player < 21:
            time.sleep(1)
            player_decision = input("score < 21, enter 'hit' to play another hand or 'stand' to stop: ").lower().strip()
            while player_decision != 'hit' and player_decision != 'stand':
                player_decision = input("Wrong entry! Enter 'hit' to play another hand or 'stand' to stop: ").lower().strip()
    
    if total_score_player > 21:
        time.sleep(1)
        print("player exceeded 21, dealer wins.")
    
    else:
        time.sleep(1)
        print("Player stands. Player score: ", total_score_player)
    
        while total_score_dealer < 17:
    
            new_card_dealer = random.choice(deck)
            time.sleep(1)
            print("Dealer new hand: ", new_card_dealer)
    
            dealer_hand.append(new_card_dealer)
            print("Dealer hands:", dealer_hand)
    
            total_score_dealer = total_score_dealer + new_card_dealer
    
        time.sleep(1)
        print("Dealer score:", total_score_dealer)
    
        if total_score_dealer > 21:
            time.sleep(1)
            print("Player wins")
    
        elif total_score_player > total_score_dealer:
            time.sleep(1)
            print("Player wins")
    
        elif total_score_player < total_score_dealer:
            time.sleep(1)
            print("Dealer wins")
    
        else:
            time.sleep(1)
            print("Tie, player score = dealer score")
    
    # 4 # Final game state
    
    game_state['player_hand'] = player_hand
    game_state['total_score_player'] = total_score_player
    game_state['dealer_hand'] = dealer_hand
    game_state['total_score_dealer'] = total_score_dealer
    print("Game state: ", game_state)
```
